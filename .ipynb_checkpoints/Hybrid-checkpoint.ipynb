```python
data_dict = {}
for col in ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Valence']:
    daily_list = []
    for date in spotify['Date'].sort_values().unique():
        spotify_date = spotify[spotify["Date"]==date]
        daily_feature = (spotify_date[col]*spotify_date['Points (Ind for each Artist/Nat)'] / 200).sum()
        daily_list.append(daily_feature)
    data_dict[col] = daily_list
```

```python
df_features_daily = pd.DataFrame(data=data_dict, index=spotify['Date'].sort_values().unique())
df_features_daily.head()
```

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
df_features_daily_std = df_features_daily.copy()
df_features_daily_std[['Loudness']] = scaler.fit_transform(df_features_daily[["Loudness"]]) * 60
```

# SerialDependence

## Setup

```python
# Set Matplotlib defaults
plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True, figsize=(11, 4))
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=16,
    titlepad=10,
)
plot_params = dict(
    color="0.75",
    style=".-",
    markeredgecolor="0.25",
    markerfacecolor="0.25",
)
%config InlineBackend.figure_format = 'retina'

def lagplot(x, y=None, lag=1, standardize=False, ax=None, **kwargs):
    from matplotlib.offsetbox import AnchoredText
    x_ = x.shift(lag)
    if standardize:
        x_ = (x_ - x_.mean()) / x_.std()
    if y is not None:
        y_ = (y - y.mean()) / y.std() if standardize else y
    else:
        y_ = x
    corr = y_.corr(x_)
    if ax is None:
        fig, ax = plt.subplots()
    scatter_kws = dict(
        alpha=0.75,
        s=3,
    )
    line_kws = dict(color='C3', )
    ax = sns.regplot(x=x_,
                     y=y_,
                     scatter_kws=scatter_kws,
                     line_kws=line_kws,
                     lowess=True,
                     ax=ax,
                     **kwargs)
    at = AnchoredText(
        f"{corr:.2f}",
        prop=dict(size="large"),
        frameon=True,
        loc="upper left",
    )
    at.patch.set_boxstyle("square, pad=0.0")
    ax.add_artist(at)
    ax.set(title=f"Lag {lag}", xlabel=x_.name, ylabel=y_.name)
    return ax

def plot_lags(x, y=None, lags=6, nrows=1, lagplot_kwargs={}, **kwargs):
    import math
    kwargs.setdefault('nrows', nrows)
    kwargs.setdefault('ncols', math.ceil(lags / nrows))
    kwargs.setdefault('figsize', (kwargs['ncols'] * 2, nrows * 2 + 0.5))
    fig, axs = plt.subplots(sharex=True, sharey=True, squeeze=False, **kwargs)
    for ax, k in zip(fig.get_axes(), range(kwargs['nrows'] * kwargs['ncols'])):
        if k + 1 <= lags:
            ax = lagplot(x, y, lag=k + 1, ax=ax, **lagplot_kwargs)
            ax.set_title(f"Lag {k + 1}", fontdict=dict(fontsize=14))
            ax.set(xlabel="", ylabel="")
        else:
            ax.axis('off')
    plt.setp(axs[-1, :], xlabel=x.name)
    plt.setp(axs[:, 0], ylabel=y.name if y is not None else x.name)
    fig.tight_layout(w_pad=0.1, h_pad=0.1)
    return fig
```

## Dumy Linear Regression

```python
def make_lags(ts, lags):
    return pd.concat(
        {
            f'y_lag_{i}': ts.shift(i)
            for i in range(1, lags + 1)
        },
        axis=1)


X = make_lags(df_danceability, lags=9)
X = X.fillna(0.0)
```

# Hybrid

```python
from sklearn.metrics import mean_absolute_error

def cal_mae(y_test, y_pred):
    print("MAE: {}".format(mean_absolute_error(y_test, y_pred)))

def cal_train_test_mae(y_train, y_fit, y_test, y_pred):
    print("Train MAE: {}".format(mean_absolute_error(y_train, y_fit)))
    print("Test MAE: {}".format(mean_absolute_error(y_test, y_pred)))
```

```python
from statsmodels.tsa.deterministic import CalendarFourier, DeterministicProcess
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

from warnings import simplefilter
simplefilter('ignore')
```

## Trend

```python
# Create trend features
selected_features = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Valence']
y = df_features_daily_std[selected_features].copy()
y = y.to_period("D").sort_index()

fourier = CalendarFourier(freq='A', order=10)
dp = DeterministicProcess(
    index=y.index,
    constant=True,
    order=1,
    # YOUR CODE HERE
    seasonal=True,
    drop=True,
    additional_terms=[fourier]
)
X = dp.in_sample()  # features for the training data

#It will be easier for us later if we
# split the date index instead of the dataframe directly.
idx_train, idx_test = train_test_split(
    y.index, test_size=0.1, shuffle=False,
)
X_train, X_test = X.loc[idx_train, :], X.loc[idx_test, :]
y_train, y_test = y.loc[idx_train], y.loc[idx_test]

# Fit trend model
model = LinearRegression(fit_intercept=False)
model.fit(X_train, y_train)

# Make predictions
y_fit_trend = pd.DataFrame(
    model.predict(X_train),
    index=y_train.index,
    columns=y_train.columns,
)
y_pred_trend = pd.DataFrame(
    model.predict(X_test),
    index=y_test.index,
    columns=y_test.columns,
)
```

```python
axs = y_train.plot(
    color='0.25', figsize=(11, 15), subplots=True, sharex=True,
    title=selected_features,
)
axs = y_test.plot(
    color='0.25', subplots=True, sharex=True, ax=axs,
)
axs = y_fit_trend.plot(
    color='C0', subplots=True, sharex=True, ax=axs,
)
axs = y_pred_trend.plot(
    color='C3', subplots=True, sharex=True, ax=axs,
)
for ax in axs: ax.legend([])
```

```python
cal_train_test_mae(y_train, y_fit_trend, y_test, y_pred_trend)
```

## Seasonality

```python
from xgboost import XGBRegressor

y_resid_trend = y_train - y_fit_trend

xgb = XGBRegressor(
    n_estimators=1000,
    max_depth=10,
)
xgb.fit(X_train, y_resid_trend)

# Add the predicted residuals onto the predicted trends
y_fit_seasonal = xgb.predict(X_train) + y_fit_trend
y_pred_seasonal = xgb.predict(X_test) + y_pred_trend
```

```python
axs = y_train.plot(
    color='0.25', figsize=(11, 15), subplots=True, sharex=True,
    title=selected_features,
)
axs = y_test.plot(
    color='0.25', subplots=True, sharex=True, ax=axs,
)
axs = y_fit_seasonal.plot(
    color='C0', subplots=True, sharex=True, ax=axs,
)
axs = y_pred_seasonal.plot(
    color='C3', subplots=True, sharex=True, ax=axs,
)
for ax in axs: ax.legend([])
```

```python
cal_train_test_mae(y_train, y_fit_seasonal, y_test, y_pred_seasonal)
```

## Cycles

```python
y_resid_seasonal_train = y_train - y_fit_seasonal
y_resid_seasonal_test = y_test - y_pred_seasonal
```

```python
lagged_num = 10
df_dict_train = {}
df_dict_test = {}
for col in y_resid_seasonal_train.columns:
    y_feature_train = y_resid_seasonal_train[col]
    y_feature_test = y_resid_seasonal_test[col]
    
    X_feature_train = make_lags(y_feature_train, lags=lagged_num).dropna()
    X_feature_test = make_lags(y_feature_test, lags=lagged_num).dropna()

    df_dict_train[col] = X_feature_train
    df_dict_test[col] = X_feature_test
    
X_features_train = pd.concat({
    "{}".format(key):
    df_dict_train[key] for key in df_dict_train.keys()
}, axis=1)

X_features_test= pd.concat({
    "{}".format(key):
    df_dict_test[key] for key in df_dict_test.keys()
}, axis=1)

y_features_train = y_resid_seasonal_train[lagged_num:]
y_features_test = y_resid_seasonal_test[lagged_num:]
```

```python
model = XGBRegressor(
)

model.fit(X_features_train, y_features_train)

y_fit_cycle = model.predict(X_features_train) + y_fit_seasonal[lagged_num:]
y_pred_cycle = model.predict(X_features_test) + y_pred_seasonal[lagged_num:]
```

```python
axs = y_train.plot(
    color='0.25', figsize=(11, 15), subplots=True, sharex=True,
    title=selected_features,
)
axs = y_test.plot(
    color='0.25', subplots=True, sharex=True, ax=axs,
)
axs = y_fit_cycle.plot(
    color='C0', subplots=True, sharex=True, ax=axs,
)
axs = y_pred_cycle.plot(
    color='C3', subplots=True, sharex=True, ax=axs,
)
for ax in axs: ax.legend([])
```
