# Explore pattern of acoustics features using points as weight

```python
data_dict = {}
acoustic_features = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Valence']
for col in acoustic_features:
    daily_list = []
    for date in spotify.index.unique():
        spotify_date = spotify[spotify.index==date]
        daily_feature = (spotify_date[col]*spotify_date['Points (Ind for each Artist/Nat)'] / 200).sum()
        daily_list.append(daily_feature)
    data_dict[col] = daily_list
```

```python
df_features_daily = pd.DataFrame(data=data_dict, index=spotify.index.sort_values().unique())
df_features_daily.head()
```

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
df_features_daily_std = df_features_daily.copy()
df_features_daily_std[['Loudness']] = scaler.fit_transform(df_features_daily[["Loudness"]]) * 60
```

# Hybrid

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from statsmodels.tsa.deterministic import CalendarFourier, DeterministicProcess
```

```python
class BoostedHybrid:
    def __init__(self, model_1, model_2):
        self.model_1 = model_1
        self.model_2 = model_2
        self.y_columns = None  # store column names from fit method
        
    def fit(self, X_1, X_2, y):
        self.model_1.fit(X_1, y)
    
        y_fit = self.model_1.predict(X_1),
    
        # YOUR CODE HERE: compute residuals
        y_resid = y - y_fit
    
        # YOUR CODE HERE: fit self.model_2 on residuals
        self.model_2.fit(X_2, y_resid)
    
        # Save column names for predict method
        self.y_columns = [y.name]
        # Save data for question checking
        self.y_fit = y_fit
        self.y_resid = y_resid

    def predict(self, X_1, X_2):
        y_pred = self.model_1.predict(X_1),
    
        # YOUR CODE HERE: add self.model_2 predictions to y_pred
        y_pred += self.model_2.predict(X_2)
        
        return y_pred # long to wide
```

```python
from sklearn.preprocessing import LabelEncoder

df = spotify[spotify.index.year==2017][acoustic_features+['Points (Total)']]

# Target series
y = df["Points (Total)"]


# X_1: Features for Linear Regression
dp = DeterministicProcess(index=y.index, order=1)
X_1 = dp.in_sample()


# X_2: Features for XGBoost
X_2 = df.drop('Points (Total)', axis=1).stack() # acoustic_features
X_2.index.set_names('acoustic_features', level=len(X_2.index.names)-1, inplace=True)

# Label encoding
le = LabelEncoder()  # from sklearn.preprocessing
X_2 = X_2.reset_index('acoustic_features')
X_2['acoustic_features'] = le.fit_transform(X_2['acoustic_features'])

# Label encoding for seasonality
X_2["month"] = X_2.index.month # values are month of the year
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor

model = BoostedHybrid(
    model_1=LinearRegression(),
    model_2=XGBRegressor()
)

# YOUR CODE HERE: Fit and predict
model.fit(X_1, X_2, y)
y_pred = model.predict(X_1, X_2)

y_pred = y_pred.clip(0.0)
```
